# Importación y Tratamiento de los Datos



```python
path = '/content/drive/MyDrive/Enseñanza de la Trigonometría (respuestas) - Respuestas de formulario 1.csv'

renombre = ['ID', 'Pais', 'Provincia','Enseña_Trigonometria', 'Curso', 'Modalidad', 'Orientacion', 'Enseño',
            'A_año', 'A_conocimientosPrevios', 'A_temasClaves', 'A_enunciado', 'A_imagen', 'A_solución', 'A_otro',
            'B_año', 'B_conocimientosPrevios', 'B_temasClaves', 'B_enunciado', 'B_imagen', 'B_solución', 'B_otro',
            'C_año', 'C_conocimientosPrevios', 'C_temasClaves', 'C_enunciado', 'C_imagen', 'C_solución', 'C_otro',
            'D_año', 'D_conocimientosPrevios', 'D_temasClaves', 'D_enunciado', 'D_imagen', 'D_solución', 'D_otro',
            'E_año', 'E_conocimientosPrevios', 'E_temasClaves', 'E_enunciado', 'E_imagen', 'E_solución', 'E_otro',
            'F_año', 'F_conocimientosPrevios', 'F_temasClaves', 'F_enunciado', 'F_imagen', 'F_solución']

dataset = pd.read_csv(path, sep=',', encoding= 'UTF-8', header=0 , names= renombre)
dataset.head()
```

```python
#Tabla Final de Profesores
table_professor = dataset[['ID', 'Pais', 'Provincia','Enseña_Trigonometria', 'Curso', 'Modalidad', 'Orientacion', 'Enseño']]

#Columnas de la Tabla Final de Problemas
columnas_problemas = ['ID', 'Año', 'ConocimientosPrevios', 'TemasClaves', 'Enunciado', 'Imagen', 'Solucion']

#División de las distintas columnas del dataset original en distintas tablas de problemas
table_problemA = dataset[['ID','A_año', 'A_conocimientosPrevios', 'A_temasClaves', 'A_enunciado', 'A_imagen', 'A_solución']]
table_problemB = dataset[['ID','B_año', 'B_conocimientosPrevios', 'B_temasClaves', 'B_enunciado', 'B_imagen', 'B_solución']]
table_problemC = dataset[['ID','C_año', 'C_conocimientosPrevios', 'C_temasClaves', 'C_enunciado', 'C_imagen', 'C_solución']]
table_problemD = dataset[['ID','D_año', 'D_conocimientosPrevios', 'D_temasClaves', 'D_enunciado', 'D_imagen', 'D_solución']]
table_problemE = dataset[['ID','E_año', 'E_conocimientosPrevios', 'E_temasClaves', 'E_enunciado', 'E_imagen', 'E_solución']]
table_problemF = dataset[['ID','F_año', 'F_conocimientosPrevios', 'F_temasClaves', 'F_enunciado', 'F_imagen', 'F_solución']]

tables = [table_problemA, table_problemB, table_problemC, table_problemD, table_problemE, table_problemF]

for t in tables:
    #renombre de las columnas de t
    t.columns = columnas_problemas
    #se eliminan las filas en las que Año sea NaN
    t.dropna(subset=['Año'], inplace=True)

#Concatenación vertical de todas las tablas de tables
table_problems = pd.concat(tables, axis=0, ignore_index=True)
table_professor.set_index('ID', inplace=True)

table_problems.head()
#table_professor.head()
```

```python
#Tabla Final de datos
mergedData = pd.merge(table_professor, table_problems, on='ID')

mergedData.head()
```

Por una cuestión visual vamos a eliminar las columnas cuyo uso fue descartado en el diseño

```python
mergedData.drop(['Provincia', 'Modalidad', 'Curso'], axis=1, inplace=True)
mergedData.head()
```

# Ingreso de los Datos del Docente y Primer Filtrado

```python
#Información docente: Marca Temporal, País, Provincia, Profesor de Trigonometria, Curso más Experimentado, Modalidad de Escuela, Orientación del Curso, Temas de Trigonometría

input = [['Argentina', 'Rio Negro', 'Si', '5 (estudiantes de 16-17 años)', 'Educación Rural', 'Agricola', 'La trigonometría es esencial en la agricultura para tareas como medir terrenos, planificar cultivos y gestionar riegos. Utiliza conceptos como las razones trigonométricas, el Teorema de Pitágoras y los triángulos rectángulos para calcular distancias, alturas y pendientes. Además, la trigonometría aplicada en los cuatro cuadrantes y la circunferencia unitaria permite analizar orientaciones y movimientos cíclicos, como el crecimiento de plantas y la rotación de cultivos, optimizando así el uso de la tierra y los recursos agrícolas.']]

frameToCompare = pd.DataFrame(input, columns=table_professor.columns)

frameToCompare.drop(['Provincia', 'Modalidad', 'Curso'], axis=1, inplace=True)

frameToCompare.head()
```

Filtrado de los datos por el pais del docente ingresado. Una vez utilizado es descartado para que no intervenga en las comparaciones

```python
ciudad = frameToCompare['Pais'].iloc[0]

df_filtrado_ciudad = table_professor[table_professor['Pais'] == ciudad]

df_filtrado_ciudad.drop(['Pais','Provincia', 'Modalidad', 'Curso'], axis=1, inplace=True)

frameToCompare.drop(['Pais'], axis=1, inplace=True)

df_filtrado_ciudad.head()
```

# Filtrado Colaborativo (Usuario - Usuario)

Si bien en el filtrado colaborativo entre usuarios se hace uso de su valoracion sobre los items de personas con gustos similares, en el ejemplo planteado los usuarios solo tiene relacion con los items que carguen, por lo que solo se utilizará para obtener los usuarios similares

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()

docentes_concatenados = df_filtrado_ciudad.apply(lambda x: ' '.join(x.astype(str)), axis=1)

entrada_concatenada = frameToCompare.apply(lambda x: ' '.join(x.astype(str)), axis=1)

docentes_a_comparar = pd.concat([docentes_concatenados, entrada_concatenada])

tfidf_matrix = vectorizer.fit_transform(docentes_a_comparar)

cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])

for idx, score in enumerate(cosine_similarities[0]):
    print(f"Similitud con la fila {idx} del DataFrame: {score:.2f}")
```

El contenido de "Temas de Trigonometria" ingresado por los docentes es extenso, por lo que por mas de que el docente ingresado tenga temas en comun (en este caso con el docente de la fila 0), solo optenemos una similitud del 0.42, por lo que se decidio filtrar a los docentes con una similitud del coseno de 1/3



```python
df_filtrado_ciudad['Similitud_coseno'] = cosine_similarities[0]

df_filtrado_coseno = df_filtrado_ciudad[df_filtrado_ciudad['Similitud_coseno'] > 0.33]

df_filtrado_coseno.head()
```

Una vez obtenidos los docentes similares al ingresado solo debemos obtener lo problemas ingresados por los mismos, los cuales serán nuestras recomendaciones

```python
docentes_similares = df_filtrado_coseno.index.tolist()

problemas_a_recomendar = table_problems[table_problems['ID'].isin(docentes_similares)]

problemas_a_recomendar.head()
```

Una vez obtenidos los items a recomendar estos serán ordenados por prioridad siendo de un mayor valor problemas que cuenten con soluciones, y a su vez dentro de dicho orden problemas que cuenten con imagenes


---


*(La preselección tanto de docentes similares como de problemas a recomedar deberia ser limitada en una cantidad optima x, sin embargo al no contar con datos de prueba el resultado sería redundante)*

```python
def prioridad_solucion_imagen(dataframe):

  dataframe['Tiene_Solucion'] = dataframe['Solucion'].notna().astype(int)
  dataframe['Tiene_Imagen'] = dataframe['Imagen'].notna().astype(int)

  dataframe_prioridad = dataframe.sort_values(by=['Tiene_Solucion', 'Tiene_Imagen'], ascending=[False, False])

  dataframe_prioridad = dataframe_prioridad.drop(['Tiene_Solucion', 'Tiene_Imagen'], axis=1)

  return dataframe_prioridad

filtrado_colaborativo = prioridad_solucion_imagen(problemas_a_recomendar)

filtrado_colaborativo.head()
```

# Recomendacion Basada en Contenido

Opcionalmente, una vez que tenemos un subconjunto de datos relacionado al docente por medio de docentes similares a el, podemos usarla para realizar una recomendacion basada en contenido, siempre y cuando el docente haya ingresado un problema para que sea tomado de referencia

```python
input_problema = [['0','6 (estudiantes de 17-18 años)', 'Razones trigonométricas', 'Ángulos', '-', 'NaN', 'NaN']]

problemaToCompare = pd.DataFrame(input_problema, columns=table_problems.columns)

contenido = ['ConocimientosPrevios','TemasClaves']

def contenido_total(data):
  return data[contenido[0]] + ' ' + data[contenido[1]]

for atributo in contenido:
 problemas_a_recomendar[atributo] = problemas_a_recomendar[atributo].fillna('')

frame_a_comparar = problemas_a_recomendar[['ID'] + contenido].apply(lambda x: ' '.join(x.astype(str)), axis=1)

problema_a_comparar = problemaToCompare[['ID'] + contenido].apply(lambda x: ' '.join(x.astype(str)), axis=1)

problemas_a_comparar = pd.concat([frame_a_comparar, problema_a_comparar])

tfidf_matrix = vectorizer.fit_transform(problemas_a_comparar)

cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])

for idx, score in enumerate(cosine_similarities[0]):
    print(f"Similitud con la fila {idx} del DataFrame: {score:.2f}")
```

En este caso la informacion cargada por los docentes es mas concisa, por lo que se filtraran los problemas con una similitud del coseno de 1/2

```python
problemas_a_recomendar['Similitud_coseno'] = cosine_similarities[0]

df_filtrado_coseno = problemas_a_recomendar[problemas_a_recomendar['Similitud_coseno'] > 0.5]

recomendacion_basada_en_contenido = prioridad_solucion_imagen(df_filtrado_coseno)

recomendacion_basada_en_contenido.head()
```
